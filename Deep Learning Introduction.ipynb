## Introduction
This is an introduction to building a neural network in Python.

In this code I follow this [example](https://machinelearningmastery.com/tutorial-first-neural-network-python-keras/) available at Machine Learning Mastery site


```python
# load modules
import pandas as pd
import numpy as np
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# load data
# add columns from description
data = pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv",
                   header=None,names=['times_pregnant','plasma_glucose_concentration',
                                      'diastolic_bp_mmhg','tricep_skin_thickness','2_hour_serum_insulin',
                                     'bmi','diabetes_pedigree','age','class'])
```

```python
# check top 5 rows
data.head()
```

```python
# get data overview
data.info()
```

```python
# data summary
data.describe()
```

```python
# split data into observations & target features
X = data.iloc[:,0:8]
y = data.iloc[:,8]
```

```python
# define a model
model = Sequential()

#add nodes & define input dimension
# define activation function
model.add(Dense(12,input_dim=8,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

# compile model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

# fit the model
model.fit(X,y,epochs=150,batch_size=10)
```

```python
# evaluate model
model_loss,model_accuracy_score = model.evaluate(X,y,verbose=0)
print('Accuracy: %.2f'%(model_accuracy_score*100))
print('Loss: %.4f'%(model_loss))
```

```python
# make prediction
for i in range(5):
    prediction = model.predict(X.loc[i:i])
    print("predicted:",np.round(prediction),"actual:",y[i])
```
